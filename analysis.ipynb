**TO SEE WHICH YEAR MOST NUMBER OF ANIME AIRED**

```python
a = df.air_date.str.count('2010').sum()
b = df.air_date.str.count('2011').sum()
c = df.air_date.str.count('2012').sum()
d = df.air_date.str.count('2013').sum()
e = df.air_date.str.count('2014').sum()
f = df.air_date.str.count('2015').sum()
g = df.air_date.str.count('2016').sum()
h = df.air_date.str.count('2017').sum()
i = df.air_date.str.count('2018').sum()
j = df.air_date.str.count('2019').sum()
k = df.air_date.str.count('2020').sum()
l = df.air_date.str.count('2021').sum()
```

```python
years = np.arange(2010,2021)
number = np.array([a,b,c,d,e,f,g,h,i,j,k])
```

```python
years
```

```python
plt.figure(figsize=(12,6))
plt.bar(years, number)
plt.xticks(years)
plt.title('No. of Anime released in a single year',fontsize = 30)
plt.xlabel('Year', color = 'grey', fontsize=20)
plt.ylabel('Total Anime', color = 'darkgrey', fontsize=20)
plt.show()
```
